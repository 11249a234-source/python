```
#Calculation of Test Average
m1 = int(input("Enter marks for Test 1: "))
m2 = int(input("Enter marks for Test 2: "))
m3 = int(input("Enter marks for Test 3: "))

minimum = min(m1, m2, m3)

sum_best_two = m1 + m2 + m3 - minimum

average_best_two = sum_best_two / 2
print("Average of best two test marks is:", average_best_two)
```
