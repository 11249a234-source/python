# Task
Create a text file, write to it, read its content, append more content, and then read the updated content using Python.

## Create and write to the file

### Subtask:
Generate code to create a text file and write some initial content into it.


## Read the file

### Subtask:
Generate code to open the created file in read mode and display its content.


**Reasoning**:
Open the file in read mode, read its content, and print it.



```python
with open("my_text_file.txt", "r") as f:
  content = f.read()
  print(content)
```

## Write to the file

### Subtask:
Generate code to open the file in append mode and add more content to it.


**Reasoning**:
Open the file in append mode and add more content.



```python
with open("my_text_file.txt", "a") as f:
  f.write("\nThis is appended content.")
```

## Read the updated file

### Subtask:
Generate code to open the file again in read mode and display the updated content.


**Reasoning**:
Open the file in read mode, read its content, and print it to display the updated content.



```python
with open("my_text_file.txt", "r") as f:
  updated_content = f.read()
  print(updated_content)
```
